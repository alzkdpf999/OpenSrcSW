```python
import csv 
f=open('tmp.csv',encoding='cp949')

data=csv.reader(f)
next(data)
#일교차 가장 큰날
chantmp=-999
#최저기온이 가장 낮은날
mintmp=999
#최고기온이 가장 높은날
maxtmp=-999
maxdate=''
mindate=''
chandate=''
for row in data:
    if row[-1] == ' ': # 최고누락이면 -999저장
        row[-1] =-999
    if row[-2] == ' ':#최저기온 누락이면 999저장
        row[-2] ==999
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])
    k= row[-1]-row[-2] 
    if chantmp<k:
        chantmp=k
        chandate=row[0]
    if maxtmp<row[-1]:
        maxtmp=row[-1]
        maxdate=row[0]
    if mintmp>row[-2]:
        mintmp=row[-2]
        mindate=row[0]

f.close()
print("일교차가 가장 큰날\n 날짜:{0:s}".format(chandate))
print("최저기온이 가장 낮은날\n 날짜:{0:s}".format(mindate))
print("최고기온이 가장 높은날\n 날짜:{0:s}".format(maxdate))
```

```python
import csv
f=open('ser.csv',encoding='cp949')
data=csv.reader(f)
next(data)
d={} 
line='1호선' #호선
re=''
for row in data:

    for i in range(4,6): #4,5     
        row[i]=row[i].strip().replace(',','')
        row[i]=int(row[i])
    k=row[4]+row[5] #총 이용객 수
    
    line_temp=row[1]
    #9호선2~3단계까지만 나온다.
    if line !=line_temp :
        #values 값으로 정렬하기 
        valuesort=sorted(d.items(),key=lambda x: x[1]) #오름차순 정렬
        minvalue = valuesort[:3]
        uvaluesort=sorted(d.items(),key=lambda x: x[1],reverse=True)#내림차순
        maxvalue = uvaluesort[:3]
        
        print("{0:s}: 이용객이 가장 많은 역 순위".format(line))
        a=1
        for i in maxvalue:
            print("{0:d}위 {1:s}({2:d})".format(a,i[0],i[1]))
            a+=1
       
                
        print("{0:s}: 이용객이 가장 적은 역 순위".format(line))
        a=1
        for i in minvalue:
            print("{0:d}위 {1:s}({2:d})".format(a,i[0],i[1]))
            a+=1
       
        print("\n")
        
        line = line_temp
        d.clear()
    d[row[3]]=k #지하철명을 key, 이용객 수를 values
    
   
    

f.close()
#한번 더 안하면 9호선2~3단계만 나온다. (if문에 의해서 우이신설선에 마지막을 읽고  
#다시 읽으면 파일이 끝나기때문에 d에 저장해놓은 것을 가지고와서 다시 정렬한뒤
#출력해준다.
valuesort=sorted(d.items(),key=lambda x: x[1]) #오름차순 정렬
minvalue = valuesort[:3]
uvaluesort=sorted(d.items(),key=lambda x: x[1],reverse=True)#내림차순
maxvalue = uvaluesort[:3]
print("{0:s}: 이용객이 가장 많은 역 순위".format(line))
a=1
for i in maxvalue:
    print("{0:d}위 {1:s}({2:d})".format(a,i[0],i[1]))
    a+=1
       
print("{0:s}: 이용객이 가장 적은 역 순위".format(line))
a=1
for i in minvalue:
    print("{0:d}위 {1:s}({2:d})".format(a,i[0],i[1]))
    a+=1
d.clear()
       
```
