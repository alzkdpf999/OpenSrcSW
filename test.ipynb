```python
a="abc def"
def first(strs,key):
    newstr=""
    for i in strs:
        code =ord(i)
        newstr+=chr(code+key)
    return newstr
def second(strs,key):
    newstr=""
    List=strs.split()
    List.reverse()
    newstr=" ".join(List)
    return first(newstr,key)
def third(strs,key):
    newstr=first(strs,key*-1)
    List=newstr.split()
    List.reverse()
    newstr=" ".join(List)
    return newstr

v=second(a,1)
d=third(v,1)
print(v)
print(d)
```

```python
#서울의 최고기온이 가장 높았던 날은 언제이고 몇 도 일까?
import csv
f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)
mx_temp= -999
date=''
for row in data:
    if row[-1] ==' ': #누락이면 -999 저장
        row[-1]=-999 
    row[-1]=float(row[-1])
    if  mx_temp < row[-1]:
        mx_temp=row[-1]
        date=row[0]
f.close()
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은 {0:s}로 {1:.2f}도 였습니다.'.format(date,mx_temp))
    
        
```

```python
#유임/무임 최대 비율?
import csv
f=open('subwayfee.csv',encoding='cp949')
data=csv.reader(f)
next(data)
mx=0 #가장 높은 비율 저장
rate=0 #비율 저장
for row in data:
    for i in range(4,8): #4,5,6,7
        row[i] = row[i].strip().replace(',', '')
        row[i]= int(row[i])
   
    if row[6]!=0:
        rate= row[4]/row[6]
        if mx < rate:
            mx=rate
            print(row,round(rate,2))
f.close()
print(round(mx,2))
```

```python
#유임 승차 비율이 가장 높은 역 유임 / (유임+무임)
import csv
f= open('subwayfee.csv',encoding='cp949')
data=csv.reader(f)
next(data)
mx=0#가장 높은 비율
rate=0# 비율
station=''#역 이름
for row in data:
    for i in range(4,8): # 4,5,6,7
        row[i]=row[i].strip().replace(',','')
        row[i]=int(row[i])
    if (row[4]+row[6])!=0:
        rate= row[4]/(row[4]+row[6])
        if mx<rate:
            station=row[3]+ ' '+ row[1]
            mx=rate
f.close()
#print('{0:s}{1:f}'.format(station,mx)) 지축 일산선 1.000000
print(station,round(mx*100,2))
```

```python
#유무임 승하차 인원이 가장 많은 역
import csv
f=open('subwayfee.csv',encoding='cp949')
data=csv.reader(f)
next(data)
mx=[0]*4 #=[0,0,0,0]
station=['']*4 #=['','','','']
label=['유임승차','유임하차','무임승차','무임하차']
for row in data:
    for i in range(4,8):
        row[i]=row[i].strip().replace(',','')
        row[i]=int(row[i])
        if row[i]>mx[i-4]:
            mx[i-4]=row[i]
            station[i-4]=row[3]+' '+ row[1]
for i in range(4):
    print(label[i]+' : '+station[i],mx[i])
```
